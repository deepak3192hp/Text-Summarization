```python
from transformers import pipeline

pipe = pipeline("summarization", model="Falconsai/text_summarization")
```

```python
def summarize_text(input_text):
    # Check if input text is provided
    if not input_text:
        return "Please enter some pdf file to summarize."
    
   # Generate the summary
    reader=  PdfReader(input_text)
    number_of_pages = len(reader.pages)
    text=""
    for page in reader.pages:
        text += page.extract_text()

    
    # Generate the summary
    summary =pipe(text , max_length=300, min_length=0, do_sample=False)
    
    return summary
```

```python
summarize_text(r"C:\Users\hp\Documents\Training\DEEPAK_DTU_MBA_BUSINESS_ANALYTICS_May_2024.pdf")
```

```python
import gradio as gr
```

```python
demo = gr.Interface(
    fn=summarize_text,
    inputs=gr.File(
        label="Upload a PDF document",
        file_types=[".pdf"],
        file_count="single"
    ),
    outputs=gr.Textbox(label="Summary"),
    title="PDF Summarization App",
    description="This application uses a Hugging Face model to generate a concise summary of your PDF document."
)

# Launch the Gradio application
if __name__ == "__main__":
    demo.launch()
```
